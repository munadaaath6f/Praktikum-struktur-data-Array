```python
# Program Python untuk mendemonstrasikan implementasi queue menggunakan list
queue = []

# Fungsi append() untuk menambahkan elemen ke dalam queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Initial queue')
print(queue)

# Fungsi pop(0) untuk menghapus elemen dari queue dengan urutan FIFO
print('\nElemen yang dihapus dari queue:')
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('\nQueue setelah menghapus elemen')
print(queue)

# Menghilangkan komentar pada print(queue.pop(0)) akan menyebabkan IndexError karena queue sudah kosong
```

```python
# Program Python untuk mendemonstrasikan implementasi queue menggunakan collections.deque
from collections import deque

queue = deque()

# Fungsi append() untuk menambahkan elemen ke dalam queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Queue awal:')
print(queue)

# Fungsi popleft() untuk menghapus elemen dari queue dengan urutan FIFO
print("\nElemen yang dihapus dari queue")
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print("\nQueue setelah menghapus elemen")
print(queue)

# Menghilangkan komentar pada print(queue.popleft()) akan menyebabkan IndexError karena queue sudah kosong
```

```python
# Program Python untuk mendemonstrasikan implementasi queue menggunakan modul queue
from queue import Queue

# Inisialisasi queue dengan ukuran maksimum 5
qu = Queue(maxsize=5)

# qsize() menampilkan jumlah elemen yang ada di dalam queue
print(qu.qsize())

# Fungsi put() untuk menambahkan elemen ke dalam queue
qu.put('a')
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o')

print("Penuh: ", qu.full())
print("Ukuran: ", qu.qsize())

# Fungsi get() untuk menghapus elemen dari queue
print("\nElemen yang dihapus dari queue")
print(qu.get())
print(qu.get())
print(qu.get())

print("\nKosong: ", qu.empty())
print("Penuh: ", qu.full())
```

```python
class Queue:
    def __init__(self):
        self.queue = list()

    def addtoqu(self, dataval):
        # Metode insert untuk menambahkan elemen
        if dataval not in self.queue:
            self.queue.insert(0, dataval)
            return True
        return False

    # Metode pop untuk menghapus elemen
    def removefromq(self):
        if len(self.queue) > 0:
            return self.queue.pop()
        return ("Tidak ada elemen dalam Queue!")

    def size(self):
        return len(self.queue)

TheQueue = Queue()
print(TheQueue.addtoqu("Jan"))
print(TheQueue.addtoqu("Feb"))
print(TheQueue.addtoqu("March"))
print(TheQueue.addtoqu("April"))
print(TheQueue.size())
print(TheQueue.removefromq())
print(TheQueue.removefromq())
print(TheQueue.size())
```
