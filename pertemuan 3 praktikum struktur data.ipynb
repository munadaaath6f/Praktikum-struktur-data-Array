```python
# Stack dengan List

# Program Python untuk mendemonstrasikan implementasi stack menggunakan list
stack = []

# Fungsi append() untuk menambahkan elemen ke dalam stack
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Stack awal')
print(stack)

# Fungsi pop() untuk mengeluarkan elemen dari stack dengan urutan LIFO
print('\nElemen yang dikeluarkan dari stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack setelah elemen dikeluarkan:')
print(stack)
```

```python
# Stack dengan collections.deque

# Program Python untuk mendemonstrasikan implementasi stack menggunakan collections.deque
from collections import deque

stack = deque()

# Fungsi append() untuk menambahkan elemen ke dalam stack
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print("Stack awal:")
print(stack)

# Fungsi pop() untuk mengeluarkan elemen dari stack dengan urutan LIFO
print("\nElemen yang dikeluarkan dari stack:")
print(stack.pop())
print(stack.pop())
print(stack.pop())

print("\nStack setelah elemen dikeluarkan:")
print(stack)
```

```python
# Stack dengan queue.LifoQueue

# Program Python untuk mendemonstrasikan implementasi stack menggunakan modul queue
from queue import LifoQueue

# Inisialisasi stack
stack = LifoQueue(maxsize=5)

# qsize() menunjukkan jumlah elemen dalam stack
print(stack.qsize())

# Fungsi put() untuk menambahkan elemen ke dalam stack
stack.put('a')
stack.put('i')
stack.put('u')
stack.put('e')
stack.put('o')

print("Penuh: ", stack.full())
print("Ukuran: ", stack.qsize())

# Fungsi get() untuk mengeluarkan elemen dari stack dengan urutan LIFO
print("\nElemen yang dikeluarkan dari stack")
print(stack.get())
print(stack.get())
print(stack.get())

print("\nKosong: ", stack.empty())
```

```python
# Stack dengan Single Linked List

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Stack:
    # Inisialisasi stack dengan dummy node untuk memudahkan penanganan kasus batas
    def __init__(self):
        self.head = Node("head")
        self.size = 0

    # Mengembalikan representasi string dari isi stack
    def __str__(self):
        cur = self.head.next
        out = ""
        while cur:
            out += str(cur.value) + "->"
            cur = cur.next
        return out[:-2]

    # Mengembalikan jumlah elemen dalam stack
    def getSize(self):
        return self.size

    # Mengecek apakah stack kosong
    def isEmpty(self):
        return self.size == 0

    # Mengembalikan nilai elemen teratas tanpa menghapusnya
    def peek(self):
        if self.isEmpty():
            raise Exception("Tidak dapat mengintip stack yang kosong")
        return self.head.next.value

    # Menambahkan elemen baru ke atas stack
    def push(self, value):
        node = Node(value)
        node.next = self.head.next
        self.head.next = node
        self.size += 1

    # Menghapus dan mengembalikan elemen teratas dari stack
    def pop(self):
        if self.isEmpty():
            raise Exception("Tidak dapat menghapus dari stack yang kosong")
        remove = self.head.next
        self.head.next = self.head.next.next
        self.size -= 1
        return remove.value

# Kode utama untuk menjalankan program
if __name__ == "__main__":
    stack = Stack()

    # Menambahkan angka 0 hingga 10 ke dalam stack
    for i in range(0, 11):
        stack.push(i)
    print(f"Stack: {stack}")

    # Menghapus 5 elemen dari stack
    for _ in range(1, 6):
        remove = stack.pop()
        print(f"Pop: {remove}")
    print(f"Stack: {stack}")
```
